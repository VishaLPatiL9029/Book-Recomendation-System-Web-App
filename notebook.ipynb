```python
books['tags'] = books['Title'] + "," + books['Author'] + "," + books['Genre'] + "," + books['SubGenre']
```

```python
books['tags'][0]
```

```python
books['tags'] = books['tags'].apply(lambda x:x.lower())
```

```python
books['tags']
```

```python
import re
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus = []

for index, row in books.iterrows():
    review = re.sub('[^a-zA-Z0-9]', ' ', row['tags'])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
df = pd.DataFrame({'Text': corpus})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 50, stop_words='english')
```

```python
vector = cv.fit_transform(df['Text']).toarray()
```

```python
cv.get_feature_names_out().tolist()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key= lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = books[books['Title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_iist = sorted(list(enumerate(distances)),reverse=True, key= lambda x: x[1])[1:6]
    
    for i in movie_iist:
        print(books.iloc[i[0]].Title)
```

```python
recommend('Data Smart')
```
